12. Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_by_position = python_data.groupby("Посада")
```

13. Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary_statistics = grouped_by_position["Зарплата на місяць"].agg(["min", "max"])
salary_statistics
```

14. Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return (row["min"] + row["max"]) / 2

salary_statistics["avg"] = salary_statistics.apply(fill_avg_salary, axis=1)
salary_statistics
```

15. Створіть описову статистику за допомогою методу describe для нового стовпчика.
16. Збережіть отриману таблицю в CSV файл

```python
avg_salary_description = salary_statistics["avg"].describe()
avg_salary_description
```

додайте від 3 до 5 графіків функцій різного типу на ваш вибір. Задайте графікам оформлення, щоб кожен графік у своїй домашній роботі чимось відрізнявся і не був схожим на інші. Можна використовувати як matplotlib, так і seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="whitegrid")
num_unique_positions = python_data["Посада"].nunique()
color_palette = sns.color_palette(
    "ch:s=.25,rot=-.25", as_cmap=False, n_colors=num_unique_positions
)
sns.boxplot(
    x="Посада",
    y="Зарплата на місяць",
    hue="Посада",
    data=python_data,
    palette=color_palette,
)
plt.title("Дистрибуція заробітної плати за посадами", fontsize=16)
plt.xlabel("Посада", fontsize=14)
plt.ylabel("Зарплата на місяць", fontsize=14)
plt.xticks(rotation=45)

plt.show()
```

```python
position_counts = python_data["Посада"].value_counts()
colors = sns.color_palette("pastel")[:num_unique_positions]

positions_pie = plt.pie(
    position_counts, autopct="%1.1f%%", pctdistance=1.17, startangle=50, colors=colors
)
plt.legend(
    positions_pie[0],
    position_counts.index,
    title="Посада",
    loc="upper right",
    bbox_to_anchor=(1, 0, 0.7, 1),
)
plt.title("Доля респондентів за посадами", fontsize=16)
plt.show()
```

```python
python_data["Загальний досвід роботи"] = pd.to_numeric(
    python_data["Загальний досвід роботи"], errors="coerce"
)
sns.lineplot(
    x="Загальний досвід роботи",
    y="Зарплата на місяць",
    data=python_data,
)

plt.title("Залежність заробітної плати від загального досвіду роботи", fontsize=16)
plt.xlabel("Загальний досвід роботи, роки")
plt.xticks(rotation=45)
plt.ylabel("Зарплата на місяць")

plt.show()
```
